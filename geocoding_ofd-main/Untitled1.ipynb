```python
import pandas as pd
import numpy as np
```

```python
bin_code = pd.read_csv('bin_add_region_code.csv', index_col=0, dtype={'bin':str})
bin_code.head()
```

```python
bin_code_ast = bin_code[bin_code['code'] == 1]
```

```python
bin_code_ast.reset_index(inplace=True,drop=True)
```

```python
bin_code_ast
```

```python
ast_true_1 = pd.read_csv('astana/astana_true_1.csv', index_col=0, dtype={'bin':str})
ast_true_1.reset_index(inplace=True,drop=True)
ast_true_1.head()
```

```python
ast_true_2 = pd.read_csv('astana/astana_true_2.csv', index_col=0, dtype={'bin':str})
ast_true_2.reset_index(inplace=True,drop=True)
ast_true_2.head()
```

```python
ast_true_3 = pd.read_csv('astana/astana_true_3.csv', index_col=0, dtype={'bin':str})
ast_true_3.reset_index(inplace=True,drop=True)
ast_true_3.head()
```

```python
astana = pd.concat([ast_true_1[['bin','lat','lon']],ast_true_2[['bin','lat','lon']],ast_true_3[['bin','lat','lon']]])
```

```python
astana['len'] = astana['bin'].str.len()
```

```python
astana['len'].value_counts()
```

```python
astana.reset_index(drop=True,inplace=True)
```

```python
astana.loc[astana['len'] == 11, 'bin'] = '0'+ astana['bin']
```

```python
astana_ofd_gis = bin_code_ast.merge(astana[['bin','lat','lon']], how='left', on='bin')
```

```python
astana_ofd_gis.dropna(inplace=True)
```

```python
import h3
def create_hexagon_clusters(data, low_level_res = 9):
    """low_level_res & high_level_res размеры гексагонов нижнего и верхнего уровня"""
    #создание кластера нижнего уровня
    data['cluster'] = [h3.geo_to_h3(row['lon'], row['lat'], low_level_res) for idx, row in data.iterrows()]
    return data
```

```python
astana_ofd_gis = create_hexagon_clusters(astana_ofd_gis)
astana_ofd_gis.head(2)
```

```python
astana_ofd_gis.groupby('cluster')['bin'].count().reset_index().to_csv('bin_ofd_cluster.csv')
```

```python
astana_ofd_gis.groupby('cluster')['bin'].count().reset_index()
```

```python
astana_ofd_gis.to_csv('astana_ofd_location.csv')
```

```python
bin_sum = pd.read_csv('bin_sum_14dec.csv', index_col=0, dtype={'KkmsRetailPlaceOrgCatalog_orgInn':str})
bin_sum.head()
```

```python
astana_ofd_sum = astana_ofd_gis.merge(bin_sum, how='left', left_on='bin', right_on='KkmsRetailPlaceOrgCatalog_orgInn')
```

```python
astana_ofd_sum.groupby('cluster')['sum'].sum().reset_index().to_csv('hex_sum_astana.csv')
```
